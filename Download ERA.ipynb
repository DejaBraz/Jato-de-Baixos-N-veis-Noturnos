**configurar o tipo de dados***

Aqui é possível selecionar qualquer variável, período e nível.

```python
#!/usr/bin/env python
import cdsapi
from datetime import datetime, timedelta
from dateutil.relativedelta import relativedelta
 
#p_levels = [str(z)  for z in ([1] + list(range(50, 1050, 50)))]
p_time = [str(time)  for time in ([1] + list(range(00, 23, 10)))]
c = cdsapi.Client()
 
 
def days_of_month(y, m):
    d0 = datetime(y, m, 1)
    d1 = d0 + relativedelta(months=1)
    out = list()
    while d0 < d1:
        out.append(d0.strftime('%Y-%m-%d'))
        d0 += timedelta(days=1)
    return out
 
for y in range(2008, 2008):
    for m in range(1,2):
        for d in days_of_month(y, m):
            c.retrieve("reanalysis-era5-pressure-levels",
                       {   "product_type": "reanalysis",
                           "variable": "temperature",
                           "pressure_level": "850",
                           "date": d,
                           "area": [20, -110, -60, -10 ], # North, West, South, East. Brasil
                           "time":p_time,
                           "format": "netcdf"
                       },
                       "ea_t_{day}.nc".format(day=d)
                       )
```
